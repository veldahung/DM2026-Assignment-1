```python
# Load data from the regularization demo CSV
df = pd.read_csv('data/NYCU_Iris.csv')
df.head()
```

## Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer

def data_preprocessing(df):
    # transform label to bi-class
    df['Species'] = df['Species'].astype(str).str.strip()
    le = LabelEncoder()
    df['Species'] = le.fit_transform(df['Species'])

    feature_cols = [c for c in df.columns if c not in ['Id', 'Species']]

    # transform string to number
    for col in feature_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    
    # TODO: Replace the missing values using “Nearest Neighbors Imputation”
    # ---------- Start your code below ----------
    
    # Step 1: Identify columns with missing values
    missing_cols = [col for col in feature_cols if df[col].isna().any()]
 
    # Step 2: Report median and std BEFORE imputation (3a)
    print("=== Before Imputation (columns with missing values) ===")
    before_stats = {}
    for col in missing_cols:
        median = df[col].median()
        std    = df[col].std()
        before_stats[col] = {'median': median, 'std': std}
        print(f"  {col:<30} | median: {median:.4f} | std: {std:.4f}")
 
    # Step 3: Apply KNN Imputation
    imputer = KNNImputer(n_neighbors=5)
    df[feature_cols] = imputer.fit_transform(df[feature_cols])
 
    # Step 4: Report median and std AFTER imputation
    print("\n=== After KNN Imputation (same columns) ===")
    for col in missing_cols:
        median = df[col].median()
        std    = df[col].std()
        print(f"  {col:<30} | median: {median:.4f} | std: {std:.4f}  "
              f"(Δmedian: {median - before_stats[col]['median']:+.4f}, "
              f"Δstd: {std - before_stats[col]['std']:+.4f})")

    # ---------- --------------------- ----------
     
    return df, feature_cols

df, feature_cols = data_preprocessing(df)
```

## Data Exploration

```python
from sklearn.feature_selection import r_regression

# TODO: Complete the 4. Data Exploration
# ── 4(a) Histogram of PetalWidthCm ──────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(df['PetalWidthCm'].dropna(), bins=20, color='steelblue', edgecolor='white')
ax.set_title('Histogram of PetalWidthCm')
ax.set_xlabel('PetalWidthCm')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
 
# ── 4(b) Feature with largest POSITIVE Pearson correlation with PetalWidthCm ─
exclude_cols = ['PetalWidthCm', 'Species', 'Id']
candidate_cols = [c for c in df.columns if c not in exclude_cols]
 
X = df[candidate_cols].values
y = df['PetalWidthCm'].values
 
pearson_r = r_regression(X, y)                         # shape: (n_features,)
corr_series = pd.Series(pearson_r, index=candidate_cols)
 
best_positive_feature = corr_series.idxmax()
print("=== 4(b) Largest Positive Correlation with PetalWidthCm ===")
print(f"  Feature : {best_positive_feature}")
print(f"  Pearson r: {corr_series[best_positive_feature]:.4f}\n")
 
# ── 4(c) Top 5 features with strongest NEGATIVE Pearson correlation ───────────
top5_negative = corr_series.nsmallest(5)
print("=== 4(c) Top 5 Strongest Negative Correlations with PetalWidthCm ===")
for feat, val in top5_negative.items():
    print(f"  {feat:<35} | r = {val:.4f}")
 
# ── 4(d) Boxplots for features found in 4(b) and 4(c) ───────────────────────
boxplot_features = [best_positive_feature] + list(top5_negative.index)
 
fig, axes = plt.subplots(1, len(boxplot_features), figsize=(16, 5))
for ax, feat in zip(axes, boxplot_features):
    ax.boxplot(df[feat].dropna(), patch_artist=True,
               boxprops=dict(facecolor='steelblue', color='navy'),
               medianprops=dict(color='orange', linewidth=2),
               whiskerprops=dict(color='navy'),
               capprops=dict(color='navy'),
               flierprops=dict(marker='o', color='red', markersize=4, alpha=0.5))
    ax.set_title(feat, fontsize=8)
    ax.set_xlabel('')
 
fig.suptitle('Boxplots of Selected Features (4b & 4c)', fontsize=12, y=1.01)
plt.tight_layout()
plt.show()
```

## Model Training

### Prepare the data

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

# normalize the data to [0,1]
for col in feature_cols:
    col_min = df[col].min()
    col_max = df[col].max()
    if col_max > col_min:
        df[col] = (df[col] - col_min) / (col_max - col_min)
    else:
        df[col] = 0.0
        
X = df[feature_cols].values.astype(float)
y = df['Species'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)

df.head()
```

### Train the model!

```python
# Use the LinearModel to fit the data

from model.linear_model import LinearModel
from model.metrics import logloss
from model.gradients import logloss_sigmoid_grad
from model.utils import *
from model.activations import sigmoid

# Model configuration
loss_fn = logloss
act_fn = sigmoid
grad_fn = logloss_sigmoid_grad

np.random.seed(seed)
model = LinearModel(dim=X_train.shape[1], is_reg=False, loss_fn=loss_fn, act_fn=act_fn, grad_fn=grad_fn)
model.fit(X_train, y_train,lr=0.1, n_iteration =10000,val_ratio=0.2, reg_type='', reg_lambda=0.0)

# print model parameters
print("Model parameters (weights):", model.W)
# sum of absolute values of weights
print("Sum of absolute values of weights:", np.sum(np.abs(model.W)))
```
